```python
#xmlを保存する
import requests
from bs4 import BeautifulSoup
import json
import os 

saved_genji_texts = os.listdir("./genji_text")
for num in range(1,55):
    if len(str(num))!=2:
        num = f"0{num}" #0~9は01,02のような形のxmlのため形を変換する
    print(F"{num=}")

    if f"{num}.json" in saved_genji_texts:
        continue

    a =requests.get(f"https://kouigenjimonogatari.github.io/tei/{num}.xml")
    soup = BeautifulSoup(a.text, 'html.parser')
    segs = soup.find_all("seg")
    title = soup.find("title").text
    texts = [k.text for k in segs]
    save_dict = {"title":title, "texts":texts}
    with open(f"./genji_text/{num}.json","w",encoding="utf-8") as f:
        json.dump(save_dict,f)
# print(f"{title=}")
# # for k in soup:
# #     print(k)
# for k in texts:
#     print(k)


```
